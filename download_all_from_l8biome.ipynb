```python
# Download the files
from tqdm import tqdm
import requests
from bs4 import BeautifulSoup
import os
```

```python
# Get the HTML content of the page
url = 'https://landsat.usgs.gov/landsat-8-cloud-cover-assessment-validation-data'  # replace with your URL
response = requests.get(url)
html_content = response.text
```

```python
html_content
```

```python
# Parse the HTML content and extract all the download links
soup = BeautifulSoup(html_content, 'html.parser')
soup
```

```python
download_links = [a['href'] for a in soup.find_all('a', href=True) if a.text]

download_links = [link for link in download_links if link.endswith('.tar.gz')]

len(download_links)
```

```python
import os

output_dir = 'D:/BIOME-dataset-images/zipped'  # replace with your output directory

# Get the list of files already present in the output directory
existing_files = os.listdir(output_dir)

# Filter the download links to exclude the files that are already present
new_download_links = [link for link in download_links if os.path.basename(link) not in existing_files]

# Total number of new files to download
total_files = len(new_download_links)

# new_download_links

# Create a progress bar
progress_bar = tqdm(total=total_files, desc='Downloading', unit='file')

for link in new_download_links:
    file_name = os.path.basename(link)
    response = requests.get(link, stream=True)
    if response.status_code == 200:
        with open(os.path.join(output_dir, file_name), 'wb') as f:
            for chunk in response.iter_content(chunk_size=1024):
                if chunk:
                    f.write(chunk)
    # Update the progress bar
    progress_bar.update(1)

# Close the progress bar
progress_bar.close()
```

```python
# top level path is 'D:/BIOME-dataset-images'
# zipped and unzipped folders are inside the top level path

# Unzip the files
import tarfile

# Define the directory containing the zip files
zipped_dir = 'D:/BIOME-dataset-images/zipped'  # replace with your directory

# Define the directory to save the unzipped files
unzipped_dir = 'D:/BIOME-dataset-images/unzipped'  # replace with your directory

# Get the list of zip files
zip_files = [f for f in os.listdir(zipped_dir) if f.endswith('.tar.gz')]

# Create a progress bar
progress_bar = tqdm(total=len(zip_files), desc='Unzipping', unit='file')

# Unzip the files
for file in zip_files:
    file_path = os.path.join(zipped_dir, file)
    with tarfile.open(file_path, 'r:gz') as tar:
        tar.extractall(unzipped_dir)
    # Update the progress bar
    progress_bar.update(1)
```
