```python
import os

output_dir = 'D:/BIOME-dataset-images/zipped'  # replace with your output directory

# Get the list of files already present in the output directory
existing_files = os.listdir(output_dir)

# Filter the download links to exclude the files that are already present
new_download_links = [link for link in download_links if os.path.basename(link) not in existing_files]

# Total number of new files to download
total_files = len(new_download_links)

# new_download_links

# Create a progress bar
progress_bar = tqdm(total=total_files, desc='Downloading', unit='file')

for link in new_download_links:
    file_name = os.path.basename(link)
    response = requests.get(link, stream=True)
    if response.status_code == 200:
        with open(os.path.join(output_dir, file_name), 'wb') as f:
            for chunk in response.iter_content(chunk_size=1024):
                if chunk:
                    f.write(chunk)
    # Update the progress bar
    progress_bar.update(1)

# Close the progress bar
progress_bar.close()
```

```python
# top level path is 'D:/BIOME-dataset-images'
# zipped and unzipped folders are inside the top level path

# Unzip the files
import tarfile

# Define the directory containing the zip files
zipped_dir = 'D:/BIOME-dataset-images/zipped'  # replace with your directory

# Define the directory to save the unzipped files
unzipped_dir = 'D:/BIOME-dataset-images/unzipped'  # replace with your directory

# Get the list of zip files
zip_files = [f for f in os.listdir(zipped_dir) if f.endswith('.tar.gz')]

# Create a progress bar
progress_bar = tqdm(total=len(zip_files), desc='Unzipping', unit='file')

# Unzip the files
for file in zip_files:
    file_path = os.path.join(zipped_dir, file)
    with tarfile.open(file_path, 'r:gz') as tar:
        tar.extractall(unzipped_dir)
    # Update the progress bar
    progress_bar.update(1)
```

```python

def process_tar_gz_and_move(tar_gz_path):
    """
    Extracts the contents of a tar.gz file and moves the inner directory to the outer directory.
    """

    # Open the tar.gz file
    tar = tarfile.open(tar_gz_path, "r:gz")

    # Define the directory to extract to (in this case, the same directory as the tar.gz file)
    extract_dir = os.path.dirname(tar_gz_path)

    # Define bc directory
    bc_dir = os.path.join(extract_dir, 'BC')

    # Extract the tar.gz file
    tar.extractall(path=extract_dir)

    # Close the tar.gz file
    tar.close()

    # Get first folder in the dir
    inner_dir_name = os.listdir(bc_dir)[0]

    # Full path to the inner directory
    inner_dir = os.path.join(bc_dir, inner_dir_name)

    # Take inner dir and move it to outer dir
    shutil.move(inner_dir, extract_dir)

    # Delete the 'BC' directory
    os.rmdir(bc_dir)
```

```python
def process_tar_gz_and_move(tar_gz_path):

    # Define the directory to extract to (in this case, the same directory as the tar.gz file)
    top_dir = os.path.dirname(tar_gz_path)

    # Get the name of the tar gz file without the extension
    file_name = os.path.basename(tar_gz_path)
    file_name = file_name.split('.')[0]

    # The name of the directory to extract to is the same as the name of the tar.gz file without the extension
    extract_dir = os.path.join(top_dir, file_name)
    # print("File: ", tar_gz_path)
    # print("Extracting to:", extract_dir)

    # Check if the directory already exists
    if not os.path.exists(extract_dir):
        # Open the tar.gz file
        print("untarring file: ", tar_gz_path)
        tar = tarfile.open(tar_gz_path, "r:gz")

        # Extract the tar.gz file
        tar.extractall(path=extract_dir)

        # Close the tar.gz file
        tar.close()

        # Define the 'bc' directory
        print("extracted to: ", extract_dir)
        bc_dir = os.path.join(extract_dir, 'BC')

        # Get first folder in the dir
        inner_dir_name = os.listdir(bc_dir)[0]

        # Full path to the inner directory
        inner_dir = os.path.join(bc_dir, inner_dir_name)

        # Take inner dir and move it to outer dir
        shutil.move(inner_dir, top_dir)

        # Delete the 'BC' directory
        os.rmdir(bc_dir)
```

```python
folder = 'C:/Users/12078/OneDrive/Documents/TESTFOLDER/etc/l8biomedownloads/unzipped'

subfolders = os.listdir(folder)

dir_reqs_list = ['_B1.TIF', '_B2.TIF', '_B3.TIF', '_B4.TIF', '_B5.TIF', '_B6.TIF', '_B7.TIF', '_B8.TIF', '_B9.TIF', '_B10.TIF', '_B11.TIF', '_BQA.TIF', '_fixedmask.hdr', '_fixedmask.img', '_Fmask4.tif', '_MTL.txt']
dir_reqs_flag = [False] * len(dir_reqs_list)

# Create a dictionary with names checked off
subfolders_dict = {}
for subfolder in subfolders:
    subfolders_dict[subfolder] = False


# Check off dir_reqs_list items for each subfolder
for subfolder in subfolders:
    for i, req in enumerate(dir_reqs_list):
        for filename in os.listdir(os.path.join(folder, subfolder)):
            if req in filename:
                dir_reqs_flag[i] = True

    # If all items in dir_reqs_flag are True, check off the subfolder in subfolders_dict
    if all(dir_reqs_flag):
        subfolders_dict[subfolder] = True

# print all subfolders that have not passed
print("Failed: ")
for subfolder in subfolders_dict:
    if not subfolders_dict[subfolder]:
        print(subfolder)

# print all subfolders that have passed
print("Passed: ")
for subfolder in subfolders_dict:
    if subfolders_dict[subfolder]:
        print(subfolder)
```
